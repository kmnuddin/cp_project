```python
i = 1
```

```python
def combine_topomaps(cluster, start, end):
    global i
    path_alpha = '../pw_band_topomaps/' + cluster + '/' + 'alpha/'
    path_beta = '../pw_band_topomaps/' + cluster + '/' + 'beta/'
    path_gamma = '../pw_band_topomaps/' + cluster + '/' + 'gamma/'
    for j in range(start, end):
        path_to_save = 'pw_band_topomaps_combined/' + str(i) + '.png'
        
        path_alpha_im = path_alpha + str(j) + '.png'
        path_beta_im = path_beta + str(j) + '.png'
        path_gamma_im = path_gamma + str(j) + '.png'
        
        im_alpha = cv2.imread(path_alpha_im)
        im_beta = cv2.imread(path_beta_im)
        im_gamma = cv2.imread(path_gamma_im)
        
        gray_alpha = cv2.cvtColor(im_alpha, cv2.COLOR_BGR2GRAY)
        gray_beta = cv2.cvtColor(im_beta, cv2.COLOR_BGR2GRAY)
        gray_gamma = cv2.cvtColor(im_gamma, cv2.COLOR_BGR2GRAY)
        
        combined_img = np.dstack((gray_alpha, gray_beta, gray_gamma))
        
        cv2.imwrite(path_to_save, combined_img)
        print(i,j)
        i += 1
        
```

```python
combine_topomaps('cluster 0', -1, 10499)
```
